```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
```

#  Data Pre-Processing

```python
filepath = os.path.join('..', '..', 'CSV for ML models', 'Summer_Olympics_Host.csv')
```

```python
df = pd.read_csv(filepath)
df.head(5)
```

```python
df_usa = df.loc[df['Country']=='USA'].reset_index(drop=True)
df_usa
```

```python
new_row = pd.Series(data={'Year':2020, 'Country':'USA', 'Athletes':340, 
                          'Host':0, 'Sport':25, 'Event':191}, name='0')
new_row
```

```python
df_usa = df_usa.append(new_row, ignore_index=True)
df_usa
```

```python
df_usa['Athletes per Event'] = df_usa['Athletes']/df_usa['Event']
```

```python
df_usa
```

```python
usa_train = df_usa.loc[df_usa['Year']<=2000]
usa_test = df_usa.loc[(df_usa['Year']>2000)]
```

```python
usa_train
```

```python
X_train = usa_train[['Year','Athletes per Event','Host']]
X_train
```

```python
y_train = usa_train["Total_Medals"]
y_train = y_train.values.reshape(-1,1)
y_train
```

```python
# Training for Gold Medals
y_train_Gold = usa_train["Gold"]
y_train_Gold = y_train_Gold.values.reshape(-1,1)


# Training for Silver Medals
y_train_Silver = usa_train["Silver"]
y_train_Silver = y_train_Silver.values.reshape(-1,1)


# Training for Bronze Medals
y_train_Bronze = usa_train["Bronze"]
y_train_Bronze = y_train_Bronze.values.reshape(-1,1)
```

```python
print(X_train.shape, y_train.shape)
```

```python
usa_test
```

```python
X_test = usa_test[['Year','Athletes per Event', 'Host']]
X_test
```

```python
y_test = usa_test["Total_Medals"]
y_test = y_test.values.reshape(-1,1)
y_test
```

```python
y_test_Gold = usa_test["Gold"]
y_test_Gold = y_test_Gold.values.reshape(-1,1)
y_test_Gold

y_test_Silver = usa_test["Bronze"]
y_test_Silver = y_test_Silver.values.reshape(-1,1)
y_test_Silver

y_test_Bronze = usa_test["Silver"]
y_test_Bronze = y_test_Bronze.values.reshape(-1,1)
y_test_Bronze
```

```python
y_test.shape
```

# Create and Train(Fit) Model

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model
```

```python
from sklearn.linear_model import LinearRegression
model_Gold = LinearRegression()


model_Silver = LinearRegression()


model_Bronze = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
model_Gold.fit(X_train, y_train_Gold)

model_Silver.fit(X_train, y_train_Silver)

model_Bronze.fit(X_train, y_train_Bronze)
```

```python
# Calculate the mean_squared_error and the r-squared value
# for the testing data
from sklearn.metrics import mean_squared_error, r2_score

# Use our model to predict a value
predicted = model.predict(X_test)
predicted
# # Score the prediction with mse and r2
# mse = mean_squared_error(y_test, predicted)
# r2 = r2_score(y_test, predicted)

# print(f"Mean Squared Error (MSE): {mse}")
# print(f"R-squared (R2 ): {r2}")
```

```python
# Calculate the mean_squared_error and the r-squared value
# for the testing data
from sklearn.metrics import mean_squared_error, r2_score

# Use our model to predict a value
predicted_Gold = model_Gold.predict(X_test)
predicted_Gold

# Use our model to predict a value
predicted_Silver = model_Silver.predict(X_test)
predicted_Silver

# Use our model to predict a value
predicted_Bronze = model_Bronze.predict(X_test)
predicted_Bronze
```

```python
# Residuals for the Training and Testing data
plt.scatter(model.predict(X_train), model.predict(X_train) - y_train, c="blue", label="Training Data");
plt.scatter(model.predict(X_test), model.predict(X_test) - y_test, c="orange", label="Testing Data")
plt.legend()
plt.hlines(y=0, xmin=y_train.min(), xmax=y_train.max())
plt.title("Residual Plot for Total Medals");
```

```python
# Residuals for the Training and Testing data
plt.scatter(model_Gold.predict(X_train), model.predict(X_train) - y_train_Gold, c="blue", label="Training Data");
plt.scatter(model_Gold.predict(X_test), model.predict(X_test) - y_test_Gold, c="orange", label="Testing Data")
plt.legend()
plt.hlines(y=0, xmin=y_train.min(), xmax=y_train.max())
plt.title("Residual Plot for Gold Medals");
```

```python
# Residuals for the Training and Testing data
plt.scatter(model_Silver.predict(X_train), model.predict(X_train) - y_train_Silver, c="blue", label="Training Data");
plt.scatter(model_Silver.predict(X_test), model.predict(X_test) - y_test_Silver, c="orange", label="Testing Data")
plt.legend()
plt.hlines(y=0, xmin=y_train.min(), xmax=y_train.max())
plt.title("Residual Plot for Silver Medals");
```

```python
# Residuals for the Training and Testing data
plt.scatter(model_Bronze.predict(X_train), model.predict(X_train) - y_train_Bronze, c="blue", label="Training Data");
plt.scatter(model_Bronze.predict(X_test), model.predict(X_test) - y_test_Bronze, c="orange", label="Testing Data")
plt.legend()
plt.hlines(y=0, xmin=y_train.min(), xmax=y_train.max())
plt.title("Residual Plot for Bronze Medals");
```
